```julia
include("age_structure_model.jl")
include("fecundity_at_age.jl")
using LinearAlgebra
using Plots
```

```julia
# single paramters 
A_max = 100 # max age
A_mature = 1 # age at maturity 
f_infty = 1.0 # maximum fecundity 
k = 0.115 # growth rate 
a0 = 1 # growth paramter
s = 0.85 # survival rate
m = 1.0 # number of migrants
z1 = 5.0 # trait value 
a_r = 2 # age when captive fish released 
a_s = 1 # age when selecitn is applied 

# vectorized paramters
fecundity = broadcast(x -> fecundity_at_age.isometric_LVB_A_mature(x,A_mature,f_infty,k,a0), 1:A_max)
survival = repeat([s],A_max)
selection = zeros(A_max)
selection[a_s] = 0.1

A = age_structure_model.build_age_matrix(A_max,fecundity,survival)
e = eigen(A)

## define popualtions 
N_init0 = abs.(e.vectors[:,argmax(real(e.values))])
N_init0[a_r] *= m
z_init0 = zeros(A_max)
z_init0[a_r] = 0
lms0 = age_structure_model.Leslie_matrix_selection(A_max,N_init0,z_init0,fecundity,survival,selection)


N_init1 = abs.(e.vectors[:,argmax(real(e.values))])
N_init1[a_r] *= m
z_init1 = zeros(A_max)
z_init1[a_r] = z1
lms1 = age_structure_model.Leslie_matrix_selection(A_max,N_init1,z_init1,fecundity,survival,selection)
```

```julia
N = 100
F_acc = zeros(N)
z = zeros(N)
n0 = zeros(N)
n1 = zeros(N)
for i in 1:N
    z[i] = lms1.z[1]
    n0[i] = lms0.N[2]
    n1[i] = lms1.N[2]
    lms1 = age_structure_model.time_step_linear!(lms1)
    lms0 = age_structure_model.time_step_linear!(lms0)

end 
Plots.plot(((n1[2:N].-n1[1:(N-1)])./n1[2:N])./((n0[2:N].-n0[1:(N-1)])./n0[2:N]))
```

```julia
a,b = 0.01,-0.5
function stage(x)
    return exp(-a-b*log(x))*x
end 
N = 20
x = 2.0
x_ls = zeros(N)
for i in 1:N
    x_ls[i] = x
    x = stage(x)
end 
Plots.plot(x_ls[1:(N-1)] .- x_ls[2:N])
```
