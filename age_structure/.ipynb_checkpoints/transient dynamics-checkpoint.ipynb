# Age stuctured sturgeon model
## Trasnient response to perturbation

This note book plots the ifluence fucntions and equilbrium diagrams for the age structured sturgeon model. It will also examine the behavior of the model linearized around the equilibrium abundace to understand the impacts of perterbations.   

```julia
include("age_structure_model.jl")
include("fecundity_at_age.jl")
include("full_trait_distribution.jl")
using LinearAlgebra
using Plots
using Roots
using Distributions
```

## Define paramters for age structured dynamics

```julia
# single paramters 
A_max = 100 # max age
A_mature = 0 # age at maturity 
f_infty = 10.0^6 # maximum fecundity 
k = 0.1 # growth rate 
a0 = 10.0 # growth paramter
s = 0.95 # survival rate


# vectorized paramters
fecundity = broadcast(x -> fecundity_at_age.isometric_LVB(x,f_infty,k,a0), 1:A_max)
survival = repeat([s],A_max)

A = age_structure_model.build_age_matrix(A_max,fecundity,survival)
print(" ")
```

## Influence function

```julia
survived = zeros(length(survival))

for i in 1:length(survival)
    survived[i] = prod(survival[1:i])
end
T_ = sum(fecundity.*survived .* collect(1:A_max) ./ sum(fecundity.*survived))
Plots.plot(fecundity.*survived .* collect(1:A_max .> A_mature))
```

## Expected life time egg production 

```julia
LEP = age_structure_model.LEP(1.0, survival, fecundity, A_max )
```

## Density dependent survival 

```julia
r = 1.5
K = 200
e_max = K*LEP
a = r/LEP
b = a/K
F(x,a,b) = a*x/(1+b*x)
f = x -> F(x,a,b) 
g = x -> F(x,a,b) - x/LEP
x = 0:10^5:e_max
Plots.plot(x, f.(x), color = "black", label = "egg - recruit")
Plots.plot!(x, x./LEP, color = "black", linestyle = :dash, label = "1/LEP")
x_eq = Roots.find_zero(g, (1, e_max))
y_eq = f(x_eq)
Plots.scatter!([x_eq],[y_eq], label = "equilibrium")
Plots.plot!(x,x./(a*LEP^2) .+(f(x_eq)-x_eq/(a*LEP^2)) , label = "linearization")
```

## Eigen value analysis 
The first three eigen values characterize the transient dynamics of the population to perterbations. One minus the first eigen value describes the resiliance of the system and the second two eigen vlaues describe the strength of oscilaitons caused by the perturbation and its interaciton with the age structure of the popuation. 

```julia
J = age_structure_model.build_age_matrix(A_max,fecundity .* 1/(a*LEP^2),survival)
vals = eigen(J)
vals.values[end-2:end]
```

## reactivity

```julia
J = age_structure_model.build_age_matrix(A_max,fecundity .* 1/(a*LEP^2),survival)
vals = eigen((J.+transpose(J))./2)
vals.values[end-2:end]
```

## Simulations with genetic effects
### parameters

```julia
min = -7.5
max = 15
dx = 0.1
grid = min:dx:max

s = 0.2
theta = 0
theta_1 = 5.0
Vle = 1.0
gradient = exp.(-s/2*(grid .- theta).^2)
d = Distributions.Normal(theta_1,sqrt(Vle))
dsn_im = pdf.(d,grid)
dsn_im = dsn_im ./ sum(dsn_im)
print(" ")
```


### correction for stabalizing selection

```julia
h(x) = 0.5/(1/x+s) + 0.5*Vle - x
Veq = Roots.find_zero(h, (0,2*Vle))
d = Distributions.Normal(theta,sqrt(Veq))
dsn = pdf.(d,grid)
dsn = dsn ./ sum(dsn)
adj = sum(dsn.*gradient)
```

### define population and dynamics

```julia

function update!(pop, R_im, dsn_im)

    # reproduction
    dsn, F = full_trait_distribution.reproduction(pop)
    

    # seleciton
    dsn, R = full_trait_distribution.selection(dsn, F, pop)
    R = R/adj
    # immigration
    dsn, R = full_trait_distribution.immigration(dsn, R, dsn_im, R_im*R)
    # dentisy dependence
    R = full_trait_distribution.recruitment(R, pop)
    # ageing!
    pop = full_trait_distribution.ageing!(pop, R, dsn)
    return pop
end
```

```julia
pop = full_trait_distribution.init_population(A_max, survival, fecundity, r ,K, theta, s, min, max, dx, Vle)

N = floor(Int,20*T_)
R_im = vcat(repeat([0],N), repeat([0.5],floor(Int,1.0*T_)), repeat([0],N))

n = zeros(length(R_im))
f = zeros(length(R_im))
W = zeros(length(R_im))
dsn = zeros(length(pop.trait))
#Plots.plot(pop.trait[:,1])
for i in 1:length(R_im)
    update!(pop, R_im[i], dsn_im)
    n[i] = sum(pop.abundance.*pop.fecundity)./sum(pop.fecundity).*pop.A_max
    f[i] = sum((pop.gradient.*pop.trait ) * (pop.abundance .* pop.fecundity) )/sum(pop.abundance .* pop.fecundity) / adj
    W[i] = full_trait_distribution.fittness(pop)
    if i == length(vcat(repeat([0],N), repeat([1.0],floor(Int,1*T_)))) + 20
        dsn = pop.trait[:,1]
    end 
end 
#Plots.plot(n./n[1])
#p1 = Plots.plot((1:N)./T_, W[(N+1):end]./W[N], color = "black")
p1 = Plots.plot((floor(Int, 9*T_):length(R_im))./T_ .- 9, 
                W[floor(Int, 9*T_):length(R_im)]./W[N],
                color = "black", 
                legend = false)
#Plots.vline!([N./T_], color = "black", linestyle = :dash, legend = false)
#p2 = Plots.plot((1:N)./T_, n[(N+1):end], color = "black")
p2 = Plots.plot((floor(Int, 9*T_):length(R_im))./T_ .- 9, 
                n[floor(Int, 9*T_):length(R_im)], 
                color = "black", 
                legend = false)
p3 = Plots.plot(dsn, color = "black")
plot(p1,p2)
```

```julia
function characterize_transient1(populaiton, R_im, dsn_im)
    # burn in 
    T_ = sum(populaiton.fecundity.*populaiton.survival .* collect(1:populaiton.A_max) ./ sum(populaiton.fecundity.*populaiton.survival))
    N = floor(Int,10*T_)
    for i in 1:N
        update!(pop, 0, dsn_im)
    end 
    W = full_trait_distribution.fittness(pop)
    min_fitness = 0
    min_abundnace = 0
    t_min_fittness = 0
    t_min_abundnace = 0
    t_05_ = 0
    t_025_ = 0
    t_0125_ = 0
    i = 0
    nls = zeros(length(R_im))
    Wls = zeros(length(R_im))
    while (t_0125_ == 0 | t_min_abundnace == 0) && i < length(R_im)
        i += 1
        update!(pop, R_im[i], dsn_im)
        nls[i] = sum(pop.abundance.*pop.fecundity)./sum(pop.fecundity).*pop.A_max
        Wls[i] = full_trait_distribution.fittness(pop) ./ W
        if i > 0.5*floor(Int,T_)
            if Wls[i] > Wls[i-1] && min_fitness == 0
                min_fitness = Wls[i-1]
                t_min_fittness = i
            end
            if t_min_fittness > 0 && (Wls[i] - min_fitness) > (1-Wls[i]) && t_05_ == 0
                t_05_ = i - t_min_fittness
            end 
            
            if t_min_fittness > 0 && (Wls[i] - min_fitness) > 3*(1-Wls[i]) && t_025_ == 0
                t_025_ = i - t_min_fittness
            end 
        
        
            if t_min_fittness > 0 && (Wls[i] - min_fitness) > 7*(1-Wls[i]) && t_0125_ == 0
                t_0125_ = i - t_min_fittness
            end 
        end 
    end 
    return min_fitness, t_min_fittness, t_05_, t_025_,t_0125_, Wls
end 
```

```julia
pop = full_trait_distribution.init_population(A_max, survival, fecundity, r ,K, theta, s, min, max, dx, Vle)

N = floor(Int,20*T_)
R_im = vcat(repeat([1.0],floor(Int,1.0*T_)), repeat([0],N))

min_fitness, t_min_fittness, t_05_, t_025_, t_0125_, Wls = characterize_transient1(pop, R_im, dsn_im)
```

```julia
Plots.plot((1:(t_min_fittness +t_0125_))./T_,Wls[1:(t_min_fittness +t_0125_)])
Plots.scatter!([t_min_fittness, t_min_fittness + t_05_, t_min_fittness + t_025_, t_min_fittness + t_0125_]./T_, 
    [min_fitness, min_fitness + 0.5(1-min_fitness), min_fitness + 0.75(1-min_fitness), min_fitness + (1-1/8)*(1-min_fitness)],
    legend = false)
```

## Why do large pulses of migrants have a smaller effect than small flows?

This could be because when there is a small
puls of migrants they are nearly all guaranteed to interbreed with natral born indivuals. This reduces the selective pressure againts the manadaptive alleles because nearly all fo the progeny of the migrants have intermidiate trait values. These intermidiate trait value experince weaker selection than the average value for the migrants because of the nonliarity in the the quadratic selection gradient. by contrast if there is a large pulse of individuals, they are more likely to breed with other migrants producing offspring with more extreme trait values that experince stronger selection removing the migrant genotypes from the popualtion more quickly.  


Another way to think about this is that larger pluses of migrants increase the variance of the geneotypes in the popualtion due to linkage disequilbirium. This allows for more rapid selection againt the new manaladapted alleles, both because incrased vairances increases the response to seleciton and becuase of the non-linar nature of the seleciton gradient.
