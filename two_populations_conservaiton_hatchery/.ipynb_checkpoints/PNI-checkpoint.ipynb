```julia
include("../quant_gen_code/normal_trait_distribution.jl")
include("two_populations_conservaiton_hatchery.jl")
using Plots
```

```julia
# Straying and maintanence of local adpatations


z = 1 # differnce in mean optimal trait values
s = 0.1 # strength of selesctions

p = 0.01 # stray rate 
p1 = p
p2 = p


```

```julia

function time_step(mu, sigma, order, z, s, p1,p2)
    Vle = 1.0
    mu_s = [-z/2,z/2] # optimal trait values for each popualtion
    sigma_s = [1.0/s,1.0/s] # variance in seleciton  for each popualtion
    if order == 1
        # straying, reproduction 
        x, mu[1], sigma[1] = normal_trait_distribution.reproduction([(1-p1),p2]/sum([(1-p1),p2]), mu, sigma, 1.0, Vle)
        x, mu[2], sigma[2] = normal_trait_distribution.reproduction([p1,(1-p2)]/sum([p1,(1-p2)]), mu, sigma, 1.0, Vle)
        # selection
        x, mu[1], sigma[1] =normal_trait_distribution.selection(x, mu[1], sigma[1], mu_s[1], sigma_s[1])
        x, mu[2], sigma[2] =normal_trait_distribution.selection(x, mu[2], sigma[2], mu_s[2], sigma_s[2])
    else
        # straying
        mu1 = mu
        mu2 = mu
        sigma1 = sigma
        sigma2 = sigma
        x1 = [(1-p1),p2]/sum([(1-p1),p2])
        x2 = [p1,(1-p2)]/sum([p1,(1-p2)])
        # selection
        print(mu1)
        print(" ")
        print(mu2)
        print("\n")
        x1[1], mu1[1], sigma1[1] =normal_trait_distribution.selection(x1[1], mu1[1], sigma1[1], mu_s[1], sigma_s[1])
        x1[2], mu1[2], sigma1[2] =normal_trait_distribution.selection(x1[2], mu1[2], sigma1[2], mu_s[1], sigma_s[1])
        x2[1], mu2[1], sigma2[1] =normal_trait_distribution.selection(x2[1], mu2[1], sigma2[1], mu_s[2], sigma_s[2])
        x2[2], mu2[2], sigma2[2] =normal_trait_distribution.selection(x2[2], mu2[2], sigma2[2], mu_s[2], sigma_s[2])
        print(mu_s)
        print("\n")
        x1 = x1/sum(x1)
        x2 = x2/sum(x2)
        print(mu1)
        print(" ")
        print(mu2)
        print("\n")
        print(x1)
        print(" ")
        print(x2)
        # reproduction   
        x, mu[1], sigma[1] = normal_trait_distribution.reproduction(x1, mu1, sigma1, 1.0, Vle)
        x, mu[2], sigma[2] = normal_trait_distribution.reproduction(x2, mu2, sigma2, 1.0, Vle)
    end 
    return mu, sigma 
end 
```

```julia
N = 1000

z = 1 # differnce in mean optimal trait values
s = 0.1 # strength of selesctions

p = 0.00 # stray rate 
p1 = p
p2 = 0.00

mu = [-z/2,z/2] # inital mean 
sigma = [Vle,Vle] # initial variance

mu_acc = zeros(N,2)
sigma_acc = zeros(N,2)
z_acc = zeros(N)
for i in 1:N
    mu_acc[i,:] = mu
    z_acc[i] = mu[2] - mu[1]
    sigma_acc[i,:] = sigma
    mu,sigma = time_step(mu, sigma, 2,z,s,p1,p2)
end
Plots.plot(mu_acc)
```

```julia
function z_diff(z0,s,p)
    N = 1000
    p1 = p
    p2 = p
    mu = [-z/2,z/2] # inital mean 
    sigma = [Vle,Vle] # initial variance
    for i in 1:N
        mu,sigma = time_step(mu, sigma, 1,z0,s,p1,p2)
    end
    return mu[2]-mu[1]
end 
s_ls = 0.01:0.01:0.5
p_ls = 0.0:0.001:0.1
z_ls = zeros(length(s_ls),length(p_ls))
i=0
for s in s_ls
    i +=1
    j = 0
    for p in p_ls
        j += 1
        z_ls[i,j] = z_diff(1,s,p)
    end
end 
Plots.heatmap(z_ls)
```
