# Conservation Hatchery Model 

This note book has some code to begin experimenting with a simple model of how domestication selection in a conservation hatchery could impact a nearby population of the same species. 

The model tracks the abundance of two populations and the mean and variance of an arbitrary trait that has effect on the fitness of the individuals. population one has low productivity and is supported by a conservation hatchery that uses brood-stock from the local population, raises the offspring in captivity and releases them into the wild. The individuals raised in the captive environment also experience selection on the trait included in the model, but the optimal value in captivity is different than the optimal value in the wild. 

Once the hatchery individuals are released, they disperse between the two populations, with the majority remaining in population one. After dispersal the populations experience density dependent mortality, brood stock is removed from population one, each population randomly mates, and the generation dies. 

The goal of this model is to understand how adaptation of hatchery reared individuals to the captive environment can effect the fitness of wild individuals in the population and in nearby populations. 

This document uses this basic model structure to investigate the effects of the conservation hatchery on population one, and the joint effect of the hatchery on both populations. As such it is not necessary to model population for some of analyses.

## Model details

To capture both the demographics and the genetic of the population we track the abundance of each wild population and the mean and variance of the breading values for the trait of interest. The size of the captive bread population is determined by the number of individuals that are released in each population. The mean and variance of the trait in this population can also be calculated from the mean and variance of the wild individuals in population one from the prior generation and so does not need to be included as a state variable.  

### Population 1
The demographic model assumes non-overlapping generations, a natural rate of increase r when the population is small in the absence of selection, and density dependent survival modeled by a Beaverton-Holt disc equation. We assume the distribution of the trait within the population is normal after reproduction, which is assumed to be random mating. Immediately after reproduction the juveniles experience trait dependent mortality. The probability of survival is set to one at the optimal trait value $\mu_{s_{p1}}$ and declines with the distance from the optimum scaled by $\sigma_{s_{p1}}$. The function for is relationship is 

$$
P(survival|x) = exp\left[-\frac{1}{2}\left(\frac{x - \mu_{s_{p1}}}{\sigma_{s_{p1}}}\right)\right].
$$

Notice that I do not model the effects of environmental or non-heritable genetic effects on the trait distribution because I assume that these effects contribute to the length scale of the selective gradient and have no other effect.    

After selection hatchery individuals are added to the population, density dependent mortality occurs, the brood stock for the hatchery is removed from the wild population and the mixture of the hatchery and wild population reproduce. A fixed fraction $s$ of the population is taken as brood stock up to a limit $s_{max}$.  

### Hatchery population
to model the demographic of the hatchery population we assume that a fixed number of individuals are released every year after selection occurs, but before density dependent mortality. After individuals are released they disperse to one of the two populations going to population one with probability $1-d$ and population 2 with probability $d$, they then experience density dependent mortality in the population they disperse to. The number that survive density dependent mortality can then join the wild individuals in their population to reproduce. The genetics are modeled by tracking a the mean and variance of the trait value. The initial trait value is set based on random mating of wild individuals from the previous generation. these values are updated to capture the effects of selection in the captive environment. This is modeled with the same functional form as selection in the wild population but a different optimal trait $\mu_{s_{h}}$ and length scale $\sigma_{s_{h}}$are used. 

### Population 2

The model for population two is quite similar to population one. I assume a fixed rate of increase in the absence of natural selection, and density dependent mortality that includes the dispersed hatchery individuals given by the Beaverton-Holt model. I assume that reproduction occurs by random mating. As before the genetic of the population are modeled by tracking the mean and the variance of the trait distribution and selection occurs immediately after birth with optimal trait $\mu_{s_{p2}}$ and length scale $\sigma_{s_{p2}}$.

### Modeling the effect of hatchery individuals on trait distribution of wild populations

To model the effect of hatchery individuals that have a distinct trait distribution mating with wild individuals I follow procedure presented in supplemental information of Springborn et al 2019 "Beyond biomass: valuing genetic diversity in natural resource management ". This method tracks the mean and variance of the trait distribution for both the wild and hatchery populations before reproduction, but assumes the trait distribution of their progeny will be normal. Formulas are given to determine the mean and variance of this distribution.

## Parameters

Although it will be useful to consider a broad range of parameters to characterize the full behavior of this model for the purposes of this document I have set the parameters to a levels that demonstrate several interesting behaviors. 

### Demographic parameters 

The demographic parameters are given by $k$ which give the carrying capacity of the population if all individuals have the optimal trait value, and $f$ which give the natural rate of increase for a population with the optimal trait value. These are set to values so that population two is not viable without support from the hatchery and population two is. I assume that 2% of population 2 is used a brood stock with a maximum value of 0.01 removed. I assume that 10% of the hatchery individuals disperse to population 2. 

### Genetic parameters
I assume that the optimal trait value for both wild population is zero and the value for hatchery population is three, and the length scale for all selection gradients are one. In addition to specifying the selection gradients I need to set a parameter $V_le$ that describes the amount of genetic diversity in a standing population, I set this value to 1.0. I am not certain what values are realistic, and presumably it is rather arbitrary given the that it is all relative the selection gradients. 


```julia
# params
x = [1.0,0.0]
mu = [0.0,0.0]
sigma = [0.25,0.25]
# parameters 
mu_s = [0.0,3.0,0.0]
sigma_s = [1.0,1.0,1.0]
k = [1.0,1.0]
f = [1.325,1.45]
V_le = 1.0
s = 0.1
prop = 0.02
max = 0.01

mod2 = two_populations_conservaiton_hatchery.population(x,mu,sigma,mu_s,sigma_s,k,f,V_le,s,prop,max)
```

## The effect of hatchery releases on population one

### Methods 
We can use this model to analyze the effect of the hatchery releases on population one in isolation because the state of population 2 does not effect the state of population 1. To begin we can consider the effect of different numbers of hatchery releases on the equilibrium state of population. The model is defined by a set of difference equations which have the general form $x_t = f(x_{t-1})$ the system is at equilibrium when $0 = f(x*) - x$. We found the equilibrium point using an algorithm called fixed point iteration. We started the algorithm with two sets of starting conditions to determine if there were multiple stable equilibrium states for any single value of hatchery releases. 

The plots output by the code chuck below show the equilibrium abundance of individuals born in the wild after selection, the mean trait value and the variance of the trait values for population one given different number of hatchery releases shown on the x axis. two lines are plotted on each graph, a dotted line representing initial values with small population sizes and a solid line representing guesses with high initial population sizes. 

### Results
There are three key observations that can be made from these figures. First, without any hatchery releases the populations not viable, but when introduced in low numbers these individuals increased the population growth rate above one for small population sizes resulting in a disproportionately large increase the abundance of the wild population relative to the numbers of releases. 

The second observation is that increasing hatchery releases increases the abundance of the wild population to a point, after which there is a bifurcation and additional hatchery releases reduce the population size. This occurs because the mean trait value sifts from values near the optimum for the wild population to an optimum near the optimum of the hatchery population, which limits the ability o the population to reproduce in the wild. 

Third, there are alternative stable states. Once the mean trait value is near the optimum for the hatchery population the number of releases can be reduced to a point that could produce a large reproducing wild population, but the distribution of traits remains near the hatchery optimum limiting the ability of the wild population to recover and locking it at low abundance. 

```julia


mod2.x = [1,1]
mod2.mu = [0.0,0.0]
mod2.sigma = [1.0,1.0]

N = 1001
x1 = zeros(N,9)
x2 = zeros(N,9)
r = collect(0.0:0.001:1).*0.3
b = 2.0
for i in 1:N
    mod2 = two_populations_conservaiton_hatchery.equilibrium_population_fpi(mod2,r[i])
    mod2, x, mu, sigma = two_populations_conservaiton_hatchery.update_populations_return_state(mod2,r[i])
    x1[i,1:3], x1[i,4:6], x1[i,7:9]= x, mu, sigma
end

for i in 1:N
    mod2 = two_populations_conservaiton_hatchery.equilibrium_population_fpi(mod2,reverse(r)[i])
    mod2, x, mu, sigma = two_populations_conservaiton_hatchery.update_populations_return_state(mod2,reverse(r)[i])
    x2[i,1:3], x2[i,4:6], x2[i,7:9]= x, mu, sigma
end


p1 = Plots.plot(r[r.<b],x1[r.<b,1],color = "black", labels = "eq1" , xlabel = "hatchery releases", ylabel = "pop 1: size")
Plots.plot!(p1,reverse(r),x2[:,1], color = "black", linestyle =:dash, labels =  "eq2")

p2 = Plots.plot(r[r.<b],x1[r.<b,4], color = "black", labels = false , xlabel = "hatchery releases", ylabel = "mean trait")
Plots.plot!(p2,reverse(r),x2[:,4], color = "black", linestyle =:dash, labels =  false)

p3 = Plots.plot(r[r.<b],x1[r.<b,7], color = "black", labels = false, xlabel = "hatchery releases", ylabel = "trait variance")
Plots.plot!(p3,reverse(r),x2[:,7], color = "black", linestyle =:dash, labels =  false)



plot(p1,p2,p3)
```

## The effects of hatchery releases on 
### Methods
To consider the effects of hatchery releases on both populations I applied the same analysis described in the previous section, but tracked the equilibrium population sizes of both wild populations and the hatchery population. 

### Results
As before there are several observationsthat can be made about the figures produced by the code chunk below. First, we see the same patters as previously described for population one (row 1). Second, we see that population two is viable without supplementation from the straying hatchery individuals, and that increasing the number released has a very limited effect it abundance, with the exception of two critical points (row 2, column 1). The first critical point occurs when the mean trait value of population one switches to a center around the hatchery optimum. At this point the equilibrium population sizes is reduced and there is a slight change in the equilibrium mean trait value and large change in variance, because strays from the hatchery now have mean traits centered around the hatchery optimum rather than the optimum for population one (row 3, column 2). The second critical point occurs when the number of hatchery releases is quite large and the mean trait value of population two switches to a center around the hatchery optimum. this leads to the collapse of the population. 

Each of these transition has an associated alternative stable equilibrium state. Both population one and population two can transition to a state with very low abundance supported by hatchery releases and strays with mean trait value centered near the hatchery optimum. The transition to this state occurs at lower number of releases for population one than population two because a larger proportion of the releases contribute to population one and because population one is less productive on it own.  This can lead to three alternative equilibrium states for some values of hatchery releases: 1) both population have mean trait near the optimum for a wild population and high abundance (solid lines in figure), 2) population one is in the hatchery dominated state and population two is in the wild dominated state (dotted line in figure), and 3) both populations are in the hatchery dominated state. 

```julia

mod2.x = [1,1]
mod2.mu = [2.5,0.0]
mod2.sigma = [0.1,0.1]


#b = 0.1975
b = 2.0
r3 = collect(0.0575:0.01:0.12)
N = length(r3)
x3 = zeros(N,9)
for i in 1:N
    mod2 = two_populations_conservaiton_hatchery.equilibrium_population_fpi(mod2,r3[i])
    mod2, x, mu, sigma = two_populations_conservaiton_hatchery.update_populations_return_state(mod2,r3[i])
    x3[i,1:3]= x
    x3[i,4:6]= mu
    x3[i,7:9]= sigma

end



r = collect(0.0:0.001:1).*0.3
p4 = Plots.plot(r[r.<b],x1[r.<b,2],color = "black", labels = false , xlabel = "hatchery releases", ylabel = "hatchery: size")
Plots.plot!(p4,reverse(r),x2[:,2], color = "black", linestyle =:dash, labels =  false)

p5 = Plots.plot(r[r.<b],x1[r.<b,5], color = "black", labels = false , xlabel = "hatchery releases", ylabel = " mean trait")
Plots.plot!(p5,reverse(r),x2[:,5], color = "black", linestyle =:dash, labels =  false)

p6 = Plots.plot(r[r.<b],x1[r.<b,8], color = "black", labels = false, xlabel = "hatchery releases", ylabel = "trait variance")
Plots.plot!(p6,reverse(r),x2[:,8], color = "black", linestyle =:dash, labels =  false)

p7 = Plots.plot(r[r.<b],x1[r.<b,3],color = "black", labels = false , xlabel = "hatchery releases", ylabel = "pop 2: size")
Plots.plot!(p7,reverse(r),x2[:,3], color = "black", linestyle =:dash, labels =  false)
Plots.plot!(p7,r3,x3[:,3], color = "black", linestyle =:dot, labels =  false)

p8 = Plots.plot(r[r.<b],x1[r.<b,6], color = "black", labels = false , xlabel = "hatchery releases", ylabel = "mean trait")
Plots.plot!(p8,reverse(r),x2[:,6], color = "black", linestyle =:dash, labels =  false)
Plots.plot!(p8,r3,x3[:,6], color = "black", linestyle =:dot, labels =  false)

p9 = Plots.plot(r[r.<b],x1[r.<b,9], color = "black", labels = false, xlabel = "hatchery releases", ylabel = "trait variance")
Plots.plot!(p9,reverse(r),x2[:,9], color = "black", linestyle =:dash, labels =  false)
Plots.plot!(p9,r3,x3[:,9], color = "black", linestyle =:dot, labels =  false)


plot(p1,p2,p3,p7,p8,p9,p4,p5,p6)

```

## Path dependent transitions

The number of hatchery releases is a bifurcation parameter, if we increase the number of individuals released from the hatchery above a critical point this can cause the mean trait value in the population to switch to the optimum value for the hatchery producing a situation where the majority of the individuals in the population are reared in the hatchery. Given that this is the case, is it possible to return to the population to the wild production dominated state only by reducing hatchery production? 

The short answer is yes, but it depends on how quickly. 

### Methods

To test if reducing hatchery releases could allow the population to recover back to the wild production dominated state I simulated series of abundance over time with, varying the levels of hatchery releases. I started these experiments by running the model for 100 generations with hatchery release set to 0.3, this allowed the population to settle into an equilibrium that was dominated by hatchery production. This point is shown in red in the figure below. 

After the population settled into the hatchery dominated equilibrium, I tested the effects of reducing the hatchery production. I reduced hatchery production to 0.05, at this point there are two alternative states the system could end up in. a Hatchery dominated state with fewer individual or the wild production dominated state. These point are marked in and orange in the figure below. 

I ran the simulations of the model. In one I reduced the hatchery releases from 0.3 to 0.05 over the course of two generations, in the second I reduced the hatchery releases to 0.05 over the course of about 30 generations and in the third I increased the hatchery releases to 2.0 for one generation and then reduced them to 0.05 the next. 



```julia

Plots.scatter!(p1,[0.3],[0.052], label = false, color = "red")
Plots.scatter!(p1,[0.05],[0.2275], label = false, color = "green")
Plots.scatter!(p1,[0.05],[0.015], label = false, color = "orange")

plot(p1)
```

### Results

The next three code chunks show the time series of abundance, mean trait value, and hatchery release for each of the three scenarios in order. The time series are broken into three regions marked by the color of the background. 

In experiment one, after the hatchery releases are reduced, the population declines precipitously and then increase to a new equilibrium. Simultaneously the mean trait value drops immediately after releases stop to an intermediate value and then slowly evolves towards a value close to the wild population optimum. The population end up in the wild production dominated state. 

In scenario two when the hatchery releases are reduced slowly, there is a slow decline in abundance that mirrors the decline in hatchery releases. The mean trait value also declines, moving towards the wild optimum. However, it does not decline as far as it did in scenario when hatchery releases are reduced quickly, as a consequence the mean trait value remains near the hatchery optimum and population size remains small. 

The third scenario was very similar to scenario two, but the increase in hatchery releases before the value was reduced increased the population size allowing the population to recover back to the wild production dominated state more quickly than in scenario one.  



```julia

n_steps = 350
mod2.x = [0.0,0]
mod2.mu = [0.0,0.0]
mod2.sigma = [1.0,1.0]

path1 = zeros(n_steps)
path2 = zeros(n_steps)
r = vcat(repeat([0.3],100),collect(0.3:-0.2:0.1) ,repeat([0.05],1000))
for i in 1:n_steps
    path1[i] = x[1]
    path2[i] = mu[1]
    mod2, x, mu, sigma = two_populations_conservaiton_hatchery.update_populations_return_state(mod2,r[i])
    
end
rectangle(w, h, x, y) = Plots.Shape(x .+ [0,w,w,0], y .+ [0,0,h,h])
Plots.plot(1:n_steps, path1, label = "abundnace", color = "black")
Plots.plot!(1:n_steps,path2, label = "mean trait", color = "black", linestyle =:dash)
Plots.plot!(1:n_steps,r[1:n_steps], label = "releases", color = "blue")
plot!(rectangle(100,1.5,0,0), opacity=.1, color = "grey", label ="burnin")
plot!(rectangle(3,1.5,100,0), opacity=.3, color = "red", label = "change releases")
plot!(rectangle(246,1.5,103,0), opacity=.1, color = "green", label = "new equilibrium")
```

```julia
# slow reduction of 
n_steps = 350
mod2.x = [0.05,0]
mod2.mu = [1.25,0.0]
mod2.sigma = [0.68,1.0]

path1 = zeros(n_steps)
path2 = zeros(n_steps)
r = vcat(repeat([0.3],100),collect(0.3:-0.005:0.05), repeat([0.05],1000))
for i in 1:n_steps
    path1[i] = x[1]
    path2[i] = mu[1]
    mod2, x, mu, sigma = two_populations_conservaiton_hatchery.update_populations_return_state(mod2,r[i])
    
end

Plots.plot(1:n_steps, path1, label = "abundnace", color = "black",legend=:right)
Plots.plot!(1:n_steps,path2, label = "mean trait", color = "black", linestyle =:dash)
Plots.plot!(1:n_steps,r[1:n_steps], label = "releases", color = "blue")
plot!(rectangle(101,1.5,0,0), opacity=.1, color = "grey")
plot!(rectangle(51,1.5,102,0), opacity=.3, color = "red")
plot!(rectangle(197,1.5,153,0), opacity=.1, color = "green")
```
