```julia
using FFTW
using Distributions
using Plots
using DSP
```

```julia
function dynamics_Normal!(pop, pop_flow)
    # selection
    pop = Normal_population.selection!(pop)
    #pop_flow = Normal_population.selection!(pop_flow)
    # print(pop_flow.N)
    # reproduction
    pop = Normal_population.reproduction!(pop, pop_flow)
    # dentisy dependence 
    pop = Normal_population.density_dependence!(pop)
    
    return pop
end 


function equilibrium_Normal!(pop, m, diff)
    pop_flow = Normal_population.population(m,0.5,1.0,s,Vle,2,[diff,Vle])
    for i in 1:500
        dynamics_Normal!(pop,pop_flow)
    end
    return pop
end 
```

```julia

function dynamics!(pop, dsn_flow,N_flow)
    
    # flow
    FFT_population.flow!(pop, dsn_flow,N_flow)
    # selection
    FFT_population.selection!(pop)
    # reproduction
    FFT_population.reproduction!(pop)
    # dentisy dependence 
    FFT_population.density_dependence!(pop)
    
end 


function equilibrium!(pop, dsn_flow,N_flow)
    for i in 1:500
        dynamics!(pop,dsn_flow,N_flow)
    end
end 
```

```julia

x = 0:0.5:10
eq = zeros(length(x))
eqv = zeros(length(x))
eqN = zeros(length(x))
m = 0.2
i = 0
for diff in x
    i +=1
    pop1, dsn_flow = FFT_population.init_pop(diff,0.5,1,0.1)
    equilibrium!(pop1, dsn_flow,m)
    eq[i] = sum(pop1.grid .* pop1.values)
    eqv[i] = sum(pop1.grid.^2 .* pop1.values) - eq[i]^2
end
Plots.plot(x,eq)
Plots.plot!(x,eqv)
Plots.plot!(x,eqN)
```
