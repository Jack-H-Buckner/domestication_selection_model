```julia
s11_w = 0.90
s22_w = 0.01
s12_w = 0.90#s11_w/2 + s22_w/2


s11_h = 0.01
s22_h = 0.90
s12_h = 0.01#s11_h/2 + s22_h/2

f = 1.2
k = 1.0
function simple_mod(x,n_hatchery)
    N = x[1]
    p = x[2]
    
    # reproduction
    N = N*f
    N11 = N * p^2
    N12 = N * 2*p*(1-p)
    N22 = N * (1-p)^2
    
    # selection 
    N11_w = N11  * s11_w
    N12_w = N12  * s12_w
    N22_w = N11  * s22_w
    

    N11_h = n_hatchery* p^2 * s11_h
    N12_h = n_hatchery * 2*p*(1-p) * s12_h
    N22_h = n_hatchery * (1-p)^2 * s22_h
    
    # density dependence
    h = (f-1)/k * (N11_w+N12_w+N22_w+N11_h+N12_h+N22_h)
    N11_w = N11_w/(1 +h)
    N12_w = N12_w/(1 +h)
    N22_w = N22_w/(1 +h)
    
    N11_h = N11_h/(1 +h)
    N12_h = N12_h /(1 +h)
    N22_h = N22_h/(1 +h)
    
    # get N and p
    N = N11_w+N12_w+N22_w+N11_h+N12_h+N22_h
    p = (2*N11_w + N12_w + 2*N11_h + N12_h) / (2*N) 
    return [N,p]
end 


function get_eq(f_hatchery)
    x = [0.1,0.01]
    for i in 1:1000
        x = simple_mod(x,f_hatchery)
    end 
    
    return x
end 
```

```julia
x = 0:0.005:2
abudnace(f) = get_eq(f)[1]
Plots.plot(x,abudnace.(x))
abudnace(f) = get_eq(f)[2]
Plots.plot!(x,abudnace.(x))
```
