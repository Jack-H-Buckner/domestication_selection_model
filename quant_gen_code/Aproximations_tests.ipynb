```julia
using FFTW
using Distributions
using Plots
using DSP
```

```julia
d = Distributions.Normal(0,1)
N = 10000
x = -5:(30/N):25
x = x
dsn = pdf.(d,x)
dsn = dsn./sum(dsn)
Plots.plot(x,dsn)
```

```julia

v1 = DSP.conv(dsn, dsn)[1:2:(2*N+1)]
v1 = v1./sum(v1)

Plots.plot(x,dsn)
Plots.plot!(x,v1)
```

```julia
x_e = -1.0:(30/N):1.0
dsn_e = pdf.(Distributions.Normal(0,0.5), x_e)
N_e = length(x_e)
dsn_e = dsn_e./sum(dsn_e)
Plots.plot(v1)
m = convert(Int64,floor(N_e/2))
Plots.plot(x,v1)
Plots.plot!(x,DSP.conv(v1, dsn_e)[(m+1):(N+m+1)])
```
