```julia
using Plots
```

```julia
x = -4:0.05:6
y = exp.(-1/2 .* (x.+0.5) .^ 2)
Plots.plot(x,y, color = "black", legend = false, ylim = [0,1.2], linewidth = 2, ticks = false)
z = 0.5 .*exp.(-1/2 .* (x .- 3 ).^ 2)
Plots.plot(x,z.+y, color = "black", legend = false, ylim = [0,1.2], linewidth = 2)
```
